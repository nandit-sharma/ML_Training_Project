##Missing values

```python
d = sns.load_dataset("titanic")
d
```

```python
d = pd.DataFrame({"Name": ['Nandit', 'Vivek', 'Kanan', 'Vikrant', 'Atul', 'Mittal'], "Age": [19, 22, None, 19, None, 18], "Gender": ["M", "M", "F",None, "M", None]})
print(d)
print("\n")
dn = d.fillna(value = {"Age": d["Age"].mean(), "Gender": "M"})
print(dn)
print("\n")
```

```python
d = pd.DataFrame({"Name": ['Nandit', 'Vivek', 'Kanan', 'Vikrant', 'Atul', 'Mittal'], "Age": [19, 22, None, 19, None, 18], "Gender": ["M", "M", "F",None, "M", None]})
print(d)
print("\n")
dn = d.fillna(method = "ffill")
print("Front Fill")
print(dn)
print("\n")
print("Back Fill")
dn = d.fillna(method = "bfill")
print(dn)
print("\n")
print("Linear Fill")
dn = d.interpolate(method = "linear")
print(dn)
print("\n")
```

```python
import pandas as pd
import sklearn
d1 = pd.read_csv("Data.csv")

la = sklearn.preprocessing.LabelEncoder()
d1["State_Code"] = la.fit_transform(d1["State"])
o = sklearn.preprocessing.OrdinalEncoder()
d1[["State_OE"]]= o.fit_transform(d1[["State"]])

d2 = pd.get_dummies(d1["State"], dtype = int)
d = pd.concat([d1, d2], axis =1)
d
```
