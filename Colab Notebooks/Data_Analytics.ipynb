#Basic maths in NumPy

```python
import numpy as np
arr = np.array([1, 2, 3])
print("1D: ", arr)
arr1 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D: ", arr1)
```

```python
z = np.zeros((3, 3))
print(z)
o = np.ones((2, 2))
print(o)
r = np.arange(0, 10, 2)
print(r)
```

```python
r = np.arange(1, 10)
s = np.arange(10, 19)
print(r,s)
print(r.shape)
print(r.ndim)
print(r.dtype)
print(np.min(r))
```

```python
print("3 x 3...")
r = r.reshape(3, 3)
s = s.reshape(3, 3)
print(r)
print("Flattern...")
r = r.flatten()
print(r)
print("Transpose...")
print(s)
print(s.T)
```
