##Standardization & Normalization

```python
df = pd.DataFrame({"Age": [25, 30, 35,40,45,50], "Salary":[25000, 30000, 35000,40000,45000,50000] })
sc = sklearn.preprocessing.StandardScaler()
sc.fit(df)
df_s = sc.transform(df)
print(df)
print('\n')
print(df_s)

sc = sklearn.preprocessing.MinMaxScaler()
sc.fit(df)
df_n = sc.transform(df)
print('\n')
print(df_n)
```

## P C A

```python

# dimensionality Reduction
X2=np.array([[1,2,3,4],[2,3,4,5],[3,4,5,6],[4,5,6,7],[5,6,7,8],[6,7,8,9]])
Y2=np.array([1,2,3,4,5,6])
from sklearn.decomposition import PCA
sc = sklearn.preprocessing.StandardScaler()
X2=sc.fit_transform(X2)
pca=PCA(n_components=2)
pca.fit(X2)
X_pca=pca.transform(X2)
print(X_pca)#
```

## S V D

```python
X=np.array([[1,2,3,4],[2,3,4,5],[3,4,5,6],[4,5,6,7],[5,6,7,8],[6,7,8,9]])
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=2)
svd.fit(X)
Xs = svd.transform(X)
print(Xs)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import numpy as np

X2 = np.array([[1,2,3,4],
               [2,3,4,5],
               [3,4,5,6],
               [4,5,6,7],
               [5,6,7,8],
               [6,7,8,9]])

Y = np.array([1, 1, 2, 2, 3, 3])  # Only 3 classes

sc = StandardScaler()
X2 = sc.fit_transform(X2)

lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X2, Y)
X_lda = lda.transform(X2)

print(X_lda)
```

##Feature Selection

```python

data={
    'A':[1,2,3,4,5],
    'B':[1,2,3,4,5],
    'C':[1,2,3,4,5]
}
df=pd.DataFrame(data)
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=1)
df=vt.fit_transform(df)
df
```

```python
#Feature Selection: Elimination
#Variance threshold
data={
    'A':[0,0,0,0,0],
    'B':[1,2,3,4,5],
    'C':[10,10,10,10,10]
}
df=pd.DataFrame(data)
from sklearn.feature_selection import VarianceThreshold
vt=VarianceThreshold(threshold=1)
vt.fit(df)
df=vt.transform(df)
df
```
