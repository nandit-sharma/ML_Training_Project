##Mean Square Error

```python
#  MEAN SQUARED ERROR

def mean(values):
  return sum(values) / len(values)


# c = y - mx

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

def coeff(x, y):
  xm = mean(x)
  ym = mean(y)
  n= 0
  d = 0
  for i in range(len(x)):
    n+= (x[i] - xm) * (y[i] - ym)
    d += (x[i] - xm)**2
  m = n/d
  c = ym - m*xm
  return m, c

m, c = coeff(x, y)



# predicted y = mx+c

def predict(x, m, c):
  p = []

  for i in x:
    yp = m*i +c
    p.append(yp)
  return p

yp = predict(x, m, c)

# MSE = 1/n[ nEi=1(yi - ypi)^2]
def mse(n, y, yp):
  sum = 0


  for i in range(n):
    e = (y[i] - yp[i])**2
    sum += e

  m = sum/n
  return m


mse = mse(len(x), y, yp)

print('Actual y = ',y)
print('Predicted y = ',yp)
print('Mean Squared Error = ',mse)






```

```python

def gradient_descent(X, y, learning_rate=0.01, epochs=1000):
 m = 0
 b = 0
 n = len(X)

 for epoch in range(epochs):
  total_error_m= 0
  total_error_c=0

  for i in range(n):
    y_pred = m * X[i] + b
    error =  y_pred - y[i]

    total_error_m +=error*X[i]
    total_error_c+=error
  m = m - (learning_rate*(2/n)*total_error_m)
  c = b - (learning_rate*(2/n)*total_error_c)


  return m, c
m, c = gradient_descent(x, y)
yp = predict(x, m, c)
mse = mse(len(x), y, yp)
print('Mean Squared Error = ',mse)
```
